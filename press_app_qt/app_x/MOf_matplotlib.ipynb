```python
import matplotlib.pyplot as plt
import numpy as np
```

### Simple plotting

```python
plt.xkcd()
month = ['Jan', 'Feb', 'Mar', 'Apr']
traffic = [150, 200, 180, 220]


plt.plot(month, traffic, marker='o', linestyle='--', color='r')


plt.xlabel('month')
plt.ylabel('traffic')
plt.title('Monthly website traffic')

plt.grid()
plt.plot()
```

### Heatmap

```python
correlation_matrix = np.random.uniform(-1, 1, (4, 3))

plt.imshow(correlation_matrix, cmap='coolwarm', vmin=-1, vmax=1,\
            aspect='auto', origin='upper')

# Add a colorbar with the following commands
cbar = plt.colorbar()
cbar.set_label('Correlation', rotation=270, labelpad=20)

# Add labels and a title
plt.title('Correlation Matrix Heatmap')
plt.xticks(range(len(correlation_matrix)), ['Var1', 'Var2', 'Var3', 'Var4'])
plt.yticks(range(len(correlation_matrix)), ['Var1', 'Var2', 'Var3', 'Var4'])

plt.show()
```
